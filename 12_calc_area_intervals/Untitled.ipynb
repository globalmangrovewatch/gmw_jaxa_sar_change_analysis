```python
import pandas
```

```python
chngs_annual_file = "base_area_stats/gmw_v314_annual_chngs_national_stats.feather"
chngs_base96_file = "base_area_stats/gmw_v314_chng_f1996_national_stats.feather"
gmw_area_extents_file = "base_area_stats/gmw_mjr_v314_national_stats.feather"
```

```python
chngs_annual_stats = pandas.read_feather(chngs_annual_file)
```

```python
cols = chngs_annual_stats.columns
cols_rm = list()
cols_gain_area = list()
cols_loss_area = list()
for col in cols:
    if "count" in col:
        cols_rm.append(col)
    elif ("area" in col) and ("gain" in col):
        cols_gain_area.append(col)
    elif ("area" in col) and ("loss" in col):
        cols_loss_area.append(col)
print(cols_rm)
print(cols_gain_area)
print(cols_loss_area)
```

```python
chngs_annual_stats = chngs_annual_stats.drop(columns=cols_rm)
chngs_annual_stats
```

```python
mng_to_nmng_com = (100-46.24930555555556)/100
mng_to_nmng_om = (100-68.56030058651027)/100
nmng_to_mng_com = (100-43.66152423012652)/100
nmng_to_mng_om = (100-63.21779214559387)/100
```

```python
print("mng_to_nmng_om = {}".format(mng_to_nmng_om))
print("mng_to_nmng_com = {}".format(mng_to_nmng_com))
print("nmng_to_mng_om = {}".format(nmng_to_mng_om))
print("nmng_to_mng_com = {}".format(nmng_to_mng_com))
```

```python
low_chngs_annual_stats = chngs_annual_stats.copy(deep=True)
up_chngs_annual_stats = chngs_annual_stats.copy(deep=True)
```

```python
for col in cols_gain_area:
    low_chngs_annual_stats[col] = low_chngs_annual_stats[col] - (low_chngs_annual_stats[col]*nmng_to_mng_com)
    up_chngs_annual_stats[col] = up_chngs_annual_stats[col] + (up_chngs_annual_stats[col]*nmng_to_mng_om)
    
for col in cols_loss_area:
    low_chngs_annual_stats[col] = low_chngs_annual_stats[col] - (low_chngs_annual_stats[col]*mng_to_nmng_com)
    up_chngs_annual_stats[col] = up_chngs_annual_stats[col] + (up_chngs_annual_stats[col]*mng_to_nmng_om)
```
