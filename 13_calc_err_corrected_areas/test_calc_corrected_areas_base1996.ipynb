```python
import pandas
```

```python
extent_file = "base_area_stats/gmw_mjr_v314_national_stats.feather"
chngs_file = "base_area_stats/gmw_v314_chng_f1996_national_stats.feather"
```

```python
extent_stats_df = pandas.read_feather(extent_file)
chng_stats_df = pandas.read_feather(chngs_file)
```

```python
chng_stats_df
```

# Remove the columns with the pixel counts so just the area columns are left

```python
cols = extent_stats_df.columns
cols_rm = list()
cols_area = list()
for col in cols:
    if "count" in col:
        cols_rm.append(col)
    elif ("area" in col):
        cols_area.append(col)
print(cols_rm)
print(cols_area)

extent_stats_df = extent_stats_df.drop(columns=cols_rm)


cols = chng_stats_df.columns
cols_rm = list()
cols_area = list()
for col in cols:
    if "count" in col:
        cols_rm.append(col)
    elif ("area" in col):
        cols_area.append(col)
print(cols_rm)
print(cols_area)

chng_stats_df = chng_stats_df.drop(columns=cols_rm)
```

```python
chng_stats_df
```

# Remove the region and name column and sum the columns to get annual totals

```python
extent_totals_df = extent_stats_df.drop(columns=["region", "name"]).sum(axis=0)
extent_totals_df/100 # divide by 100 for km2
```

# Split the gains and losses into seperate dataframes

```python
chng_totals_df = chng_stats_df.drop(columns=["index", "region", "name"]).sum(axis=0)

gain_cols = ["2007_area_gain", "2008_area_gain", "2009_area_gain", "2010_area_gain", "2015_area_gain", "2016_area_gain", "2017_area_gain", "2018_area_gain", "2019_area_gain", "2020_area_gain"]
loss_cols = ["2007_area_loss", "2008_area_loss", "2009_area_loss", "2010_area_loss", "2015_area_loss", "2016_area_loss", "2017_area_loss", "2018_area_loss", "2019_area_loss", "2020_area_loss"]

chng_loss_totals_df = chng_totals_df.drop(gain_cols)
chng_gain_totals_df = chng_totals_df.drop(loss_cols)

chng_loss_stats_df = chng_stats_df.drop(columns=gain_cols)
chng_gain_stats_df = chng_stats_df.drop(columns=loss_cols)
```

```python
chng_loss_totals_df/100
```

```python
chng_gain_totals_df/100
```

# Define the Ommission and Commissions

```python
mng_om = (100-85.63061873343034)/100
mng_com = (100 - 89.28977298408603)/100
```

```python
print(f"mng_om = {mng_om}")
print(f"mng_com = {mng_com}")
```

```python
loss_om = (100-73.77949765590216)/100
loss_com = (100-51.42118863049095)/100
```

```python
print(f"loss_om = {loss_om}")
print(f"loss_com = {loss_com}")
```

```python
gain_om = (100-69.97109826589596)/100
gain_com = (100-50.0)/100
```

```python
print(f"gain_om = {gain_om}")
print(f"gain_com = {gain_com}")
```

# Apply to annual totals to see what the numbers are:

```python
extent_totals_corr_df = extent_totals_df + (extent_totals_df * mng_om) - (extent_totals_df * mng_com)
extent_totals_corr_df/100
```

```python
chng_loss_totals_corr_df = chng_loss_totals_df + (chng_loss_totals_df * loss_om) - (chng_loss_totals_df * loss_com)
chng_loss_totals_corr_df/100
```

```python
chng_gain_totals_corr_df = chng_gain_totals_df  + (chng_gain_totals_df * gain_om) - (chng_gain_totals_df * gain_com)
chng_gain_totals_corr_df/100
```

# Calculate the net change from the corrected gains and losses

```python
net_chng_vals = dict()

gmw_chng_years = [2007, 2008, 2009, 2010, 2015, 2016, 2017, 2018, 2019, 2020]
for chng_year in gmw_chng_years:
    net_chng_vals[f"{chng_year}_net_chng"] = chng_loss_totals_corr_df[f"{chng_year}_area_loss"] - chng_gain_totals_corr_df[f"{chng_year}_area_gain"]

net_chng_corr_sr = pandas.Series(net_chng_vals)
net_chng_corr_sr/100
```

# Create copies of the dataframes and set the region to be the index so the countries get correctly matched

```python
extent_stats_corr_test_df = extent_stats_corr_df.copy()
extent_stats_corr_test_df = extent_stats_corr_test_df.set_index("region")

chng_gain_corr_stats_test_df = chng_gain_corr_stats_df.copy()
chng_gain_corr_stats_test_df = chng_gain_corr_stats_test_df.set_index("region")
chng_loss_corr_stats_test_df = chng_loss_corr_stats_df.copy()
chng_loss_corr_stats_test_df = chng_loss_corr_stats_test_df.set_index("region")
```

# Apply change correction to calculate country extent per year

```python
gmw_chng_years = [2007, 2008, 2009, 2010, 2015, 2016, 2017, 2018, 2019, 2020]

for chng_year in gmw_chng_years: 
    extent_stats_corr_test_df[f"{chng_year}_area"] = extent_stats_corr_test_df["1996_area"] - chng_loss_corr_stats_test_df[f"{chng_year}_area_loss"] + chng_gain_corr_stats_test_df[f"{chng_year}_area_gain"]
```

# Find countries for which the area is NaN as we have no changes for these countries

```python
no_chng_rgns_df = extent_stats_corr_test_df[extent_stats_corr_test_df.isna().any(axis=1)]
no_chng_rgns_df
```

# Replace the NaN values will the 1996 area

```python
for i, row in no_chng_rgns_df.iterrows():
    print(i)
    for chng_year in gmw_chng_years: 
        #print(f"\t{chng_year}")
        extent_stats_corr_test_df.at[i, f"{chng_year}_area"] = extent_stats_corr_test_df.loc[[i]]["1996_area"]
```
